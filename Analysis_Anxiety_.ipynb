```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('anxiety_attack_dataset.csv')
print(data.head())
print(data.info())
```

```python
np.random.seed(42)
```

```python
import numpy as np
data_size = 100
data = {
    'ID':range(1, data_size + 1),
    'Age':np.random.randint(18, 65, size=data_size),
    'Gender':np.random.choice(['Male', 'Female'], size=data_size),
    'Anxiety_Level':np.random.uniform(0, 10, size=data_size)
}
```

```python
df = pd.DataFrame(data)
summary_stats = df.describe()
print(summary_stats)
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Anxiety_Level', data=df)
plt.title('Anxiety Levels by Gender')
plt.xlabel('Gender')
plt.ylabel('Anxiety Level')
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(df['Anxiety_Level'], bins=20, kde=True)
plt.title('Distribution of Anxiety Levels')
plt.xlabel('Anxiety Level')
plt.ylabel('Frequency')
plt.show()
```

```python
correlation_matrix = df.select_dtypes(include=np.number).corr()
print(correlation_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
